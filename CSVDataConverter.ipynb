```python
import numpy as np
import pandas as pd
import datetime
```

```python
# config

# time unit
# 'TIMESTAMP' or 'BAR'
TIMESTAMP, BAR = 0, 1
TIME_UNIT = BAR
```

```python
df = pd.read_csv('data/e-mini-sp500-200530/e-mini-sp500-daily.csv')

if TIME_UNIT == TIMESTAMP:
    df['Timestamp'] = (df['Date'] + '-' + df['Time']).map(lambda s: datetime.datetime.strptime(s, '%m/%d/%Y-%H:%M').timestamp())
elif TIME_UNIT == BAR:
    df['Timestamp'] = np.arange(df.shape[0])
else:
    raise ValueError(f'TIME_UNIT must be {TIMESTAMP} or {BAR}, but got {TIME_UNIT}.')

df = df.drop(['Date', 'Time', 'OI'], axis=1)
df = df.rename(columns={'Timestamp': 'Time'}) 
df.head(20)
```

```python
df.to_csv('data/compact/e-mini-sp500-daily-compact.csv', index=False)
```
